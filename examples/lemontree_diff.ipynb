# Parse SysML model from string and convert to SysML v2 API Flexo JSON

```python
sysml_model_string_1 = '''
package pkg {
   part p;
}
'''

sysml_model_string_2 = '''
package pkg {
    part p;
    part p1;
}
'''


change_payload_str, raw_json_base = convert_sysml_string_textual_to_json(sysml_model_string_1, minimal=True)
change_payload_str, raw_json_branch = convert_sysml_string_textual_to_json(sysml_model_string_2, minimal=True)

```

```python
(sysml_text, model), captured_warnings = convert_json_to_sysml_textual(raw_json_base)
print(f"Created SysML model:\n{sysml_text}")
(sysml_text, model), captured_warnings = convert_json_to_sysml_textual(raw_json_branch)
print(f"Created SysML model:\n{sysml_text}")
```

```python
with open("lt_base.json", "w", encoding="utf-8") as f:
    f.write(str(raw_json_base))
with open("lt_branch.json", "w", encoding="utf-8") as f:
    f.write(str(raw_json_branch))
```

```python
with open("lt_merged.json", "r", encoding="utf-8") as f:
    data = json.load(f)
(sysml_text, model), captured_warnings = convert_json_to_sysml_textual(data)
print(f"Created SysML model:\n{sysml_text}")
```
